Distibuição de jogadores por idade

```python
plt.figure(figsize=(15,10))
order = df['Age'].value_counts().index
sns.countplot(x='Age', data=df, order = sorted(order))
plt.xticks(rotation=90);
```

Distruibuição de jogadores por peso

```python
plt.figure(figsize=(15,10))
order = df['Weight'].value_counts().index
sns.countplot(x='Weight', data=df, order=sorted(order))
plt.xticks(rotation=90);
```

```python
plt.figure(figsize=(15,10))
order = df['Height'].value_counts().index
sns.countplot(x='Height', data=df, order=sorted(order))
plt.xticks(rotation=90);
```
